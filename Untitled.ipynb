```python
from netCDF4 import num2date, Dataset as NetCDFFile
import numpy as np
import os
import matplotlib.pyplot as plt
os.environ['PROJ_LIB'] = r'C:\Users\Maxim\Anaconda3\pkgs\proj4-5.2.0-ha925a31_1\Library\share'
from mpl_toolkits.basemap import Basemap
```

```python
nc_file = 'C:/Users/Maxim/Downloads/_grib2netcdf-atls12-a82bacafb5c306db76464bc7e824bb75-RIwsLe.nc'
nc = NetCDFFile(nc_file)

lat = nc.variables['latitude'][:]
lon = nc.variables['longitude'][:]
time = nc.variables['time'][:]
t2 = nc.variables['t2m'][:]  # 2 meter temperature
t2_units = nc.variables['t2m'].units

dates = num2date(nc.variables['time'][:],units=nc.variables['time'].units,calendar=nc.variables['time'].calendar)
nc.close()
```

```python
# Get some parameters for the Stereographic Projection
lon_0 = lon.mean()
lat_0 = lat.mean()

map = Basemap(width=5000000,height=3500000,
            resolution='l',projection='stere',\
            lat_ts=40,lat_0=lat_0,lon_0=lon_0)
```

```python
map.drawcoastlines()
map.drawstates()
map.drawcountries()
map.drawlsmask(land_color='Linen', ocean_color='#CCFFFF') # can use HTML names or codes for colors
```

```python
# Because our lon and lat variables are 1D,
# use meshgrid to create 2D arrays
# Not necessary if coordinates are already in 2D arrays.
lons, lats = np.meshgrid(lon, lat)
xi, yi = map(lons, lats)
```

```python
from ipywidgets import *
# Plot Data
def draw(timeItem):
    cs = map.pcolor(xi,yi, t2[timeItem,:,:])

    # Add Grid Lines
    map.drawparallels(np.arange(-80., 81., 10.), labels=[1,0,0,0], fontsize=10)
    map.drawmeridians(np.arange(-180., 181., 10.), labels=[0,0,0,1], fontsize=10)

    # Add Coastlines, States, and Country Boundaries
    map.drawcoastlines()
    map.drawstates()
    map.drawcountries()

    # Add Colorbar
    cbar = map.colorbar(cs, location='bottom', pad="10%")
    cbar.set_label(t2_units)

    # Add Title
    plt.title('Temperature 2 metre')
    print(dates[timeItem])
    plt.show()
```

```python
interact(draw, timeItem=(0, len(time) - 1, 1))
```
